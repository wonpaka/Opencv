```python
import requests

img_lists = list(set(img_lists))
if 'data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7' in img_lists:
    img_lists.pop(img_lists.index('data:image/gif;base64,R0lGODlhAQABAIAAAAAAAP///yH5BAEAAAAALAAAAAABAAEAAAIBRAA7'))
    
print(len(img_lists))
```

```python
# 폴더 생성

import os
 
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

createFolder(f'./image/{eng_name}')
```

```python
# 파일 입력

for i in range(len(img_lists)):
    r = requests.get(img_lists[i])
    file = open(f'./image/{eng_name}/{eng_name}_' + str(i) + '.jpg',"wb")
    file.write(r.content)
    file.close()
```
