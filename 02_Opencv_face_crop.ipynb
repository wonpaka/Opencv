# 얼굴이미지 추출

> - 일단 다음과 같은 방식으로 한장에 대해 테스트를 진행
> - 이를 활용하여 다량의 사진을 변환 시킨다.
> - color는 무시한다. (얼굴 추출에 색은 필요가 없을듯..)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
```

### 1) 초기설정

```python
# 이거만 바꿔 주면 된다.
eng_name = 'Yezi'
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# 폴더 생성
if not os.path.exists(f'./image/train_{eng_name}'):
    os.makedirs(f'./image/train_{eng_name}')
    
# Raw 데이터 개수 파악
len_files = len(os.listdir(f'./image/{eng_name}'))

#test
i = 0
```

### 2) 전체 탐지

```python
for i in range(len_files):
    # 이미지 불러오기
    image = cv2.imread(f'./image/{eng_name}/{eng_name}_{str(i)}.jpg')

    # 흑백
    grayImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # 얼굴 찾기
    faces = face_cascade.detectMultiScale(grayImage, 1.03, 5)
    
    # 얼굴이 없는 경우
    if type(faces) == tuple:
        pass
    
    #얼굴이 있다면,
    else :
        # 인식된 얼굴 개수가 하나라면
        if not faces.shape[0] -1 :
            for x,y,w,h in faces:
                cropped = grayImage[y - int(h/4):y + h + int(h/4), x - int(w/4):x + w + int(w/4)]
                # 원본크기 좌표를 넘어가면 shape는 0으로 나옴 따라서 패딩을 순차적으로 줄여줌 / 나는 두번만 진행함
                if cropped.shape[0] == 0 or cropped.shape[1] == 0 :
                    cropped = grayImage[y - int(h/8):y + h + int(h/8), x - int(w/8):x + w + int(w/8)]
                    if cropped.shape[0] == 0 or cropped.shape[1] == 0 :
                        cropped = grayImage[y:y + h, x:x + w]

        # 두개 이상이라면
        else:
            max_w= 0
            real_face = [0,0,0,0]

            for (x,y,w,h) in faces:
                if w > max_w:
                    max_w = w
                    real_face = [[x,y,w,h]]

            for x,y,w,h in real_face:
                cropped = grayImage[y - int(h/4):y + h + int(h/4), x - int(w/4):x + w + int(w/4)]
                # 원본크기 좌표를 넘어가면 shape는 0으로 나옴 따라서 패딩을 순차적으로 줄여줌 / 나는 두번만 진행함
                if cropped.shape[0] == 0 or cropped.shape[1] == 0 :
                    cropped = grayImage[y - int(h/8):y + h + int(h/8), x - int(w/8):x + w + int(w/8)]
                    if cropped.shape[0] == 0 or cropped.shape[1] == 0 :
                        cropped = grayImage[y:y + h, x:x + w]

            
        # if 에러 패스 / else 크기 통일
        # INTER_CUBIC : 바이큐빅 보간법 -> 확대
        # INTER_AREA : 영역 보간법 -> 축소
        if cropped.shape[0] != cropped.shape[1]:
            pass
        
        else :
            if cropped.shape[0] <100 or cropped.shape[1] <100:
                cropped = cv2.resize(cropped, dsize=(100, 100), interpolation=cv2.INTER_CUBIC)
            else:
                cropped = cv2.resize(cropped, dsize=(100, 100), interpolation=cv2.INTER_AREA)
            
            # 여기서 눈이 없으면 또 잘못된 추출값임
            # 눈 찾기
            eyes = eye_cascade.detectMultiScale(cropped, 1.03, 5)
            
            if type(eyes) == tuple:
                pass
            
            else :
                # 눈이 하나인 경우도 일단 제거한다.. 유의성을 위해 / 거의 데이터 반토막
                if eyes.shape[0] == 1:
                    pass
                
                else:
                    # 저장
                    cv2.imwrite(f'./image/train_{eng_name}/train_{eng_name}_{str(i)}.jpg', cropped)
                    print(f'\'train_{eng_name}_{str(i)}.jpg\' 이미지 생성 완료!')         
```

## 진행 중 발생 문제점 및 보완 방법

> - 1) 얼굴인식이 안되는 경우 검색값이 NuLL로 나타나며 tuple 형태로 반환됨, 따라서 이경우 Loop에서 제외시킴
> - 2) 얼굴인식이 되면 패딩을 주어 더 넓은 영역을 잘라내는데 이때 영역확장시 원본을 넘어가는 경우가 생김
>
>    이 경우 패딩을 순차적으로 잘라주며 그나마 최대값을 유지하려 노력함


> - cv2 공간보간법으로 확대 축소 진행
> - 오차값이 너무 많아 눈이 2개이상 존재해야 이미지를 다운받을수 있도록 진행함 (영역크기, 위치로 구체화 가능할듯)
